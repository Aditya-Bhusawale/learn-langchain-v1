## 1) Model integration with OpenAI,Groq and Google Gemini model

## 2) Streaming and Batch

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

### OpenAI model

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("gpt-4.1")      ## you can use chatopenai() also
model
```

```python
## invoke the model

response=model.invoke("hello, how are you?")
response
```

### Google Gemini model

```python
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")

from langchain.chat_models import init_chat_model
model=init_chat_model("google_genai:gemini-2.5-flask-lite")     ## you can use chatgooglegenerativeai() also
response=model.invoke("hello, how are you?")
response
```

### Groq model

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("groq:llama-3.3-70b-versatile")      ## you can use chatgroq() also
response=model.invoke("hello, how are you?")
response.content
```

### Streaming

If you mean streaming in LangChain, LangChain provides a .stream() method (and .astream() for async) that lets you receive model output token-by-token instead of waiting for the full response.

```python
for chunk in model.stream("what is AI"):
    print(chunk.text,end='',flush=True)
```

### Batch

In LangChain, batch() lets you send multiple inputs at once and get all responses back together.

```python
questions = [
    "What is AI?",
    "What is Machine Learning?",
    "What is Deep Learning?"
]

responses=model.batch(questions)

for response in responses:
    print(response)
```
